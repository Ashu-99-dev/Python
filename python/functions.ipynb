```python
# Q2. Explain the concept of function arguments and parameters in Python.

## Function arguments -> are the values that are send to the function when it is called.
def add(a,b): # here int a and b are parameters
    print(a+b);
## Function parameters-> are the variables listed inside the parentheses in the function definition.
add(2,3); # here 2 and 3 are arguments
```

```python
# Q3. What are the different ways to define and call a function in Python?

## To define a function in pyhton, you use the def keyword followed by the function name, parentheses(),and a colon : .
## In Python their are two type of function 1. non-parameter 2. parameter function
# 1. non- parameter
def my_function():
    print("Hello, World!")

# 2. parameter
def sub(a,b):
    print(a-b);

## To call a function, you use the function name followed by parentheses ().
my_function()
sub(10,5)
```

```python
# Q. What is the purpose of the `return` statement in a Python function?

## The return statement is used to exit a function and return a value to the caller.
## When the return statement is executed, the function terminates immediately, and the value specified after the return keyword is passed back to the part of the program that called the function.

def add(a,b):
    return a+b;

add(2,3)
```

```python
# Q5. What are iterators in Python and how do they differ from iterables?

##  An iterable is any Python object capable of returning its members one at a time, allowing it to be iterated over in a loop.

my_list = [1, 2, 3, 4, 5]

## An iterator is an object that repersents a stream of data. It implements the iterator protocol.
## You can obtain an iterator froman iterable using the built in iter() function.

my_iterator = iter(my_list)

print(next(my_iterator))
print(next(my_iterator))
print(next(my_iterator))

```

```python
# Q6. Explain the concept of generators in Python and how they are defined.

# Generators are a special type of iterator that produce a sequence of values on demand, rather than storing them all in memory at once.
# They defined using functions, but instead of using the return statement, they use the yeild keyword.
# When a generator function is called, it doesnot execute the function body immediately. Instesd, it returns a generator object.
# This object can iterates over to retrive the values yielded by the function.

def my_generator(n):
  for i in range(n):
    yield i

gen = my_generator(5)
for i in gen:
  print(i)
```

```python
# Q7. What are the advantages of using generators over regular functions?

## Their are many advantages of using generators over regular functions some of are listed below.
## Memory efficiency due to on-demand value generation
## Lazy evaluation for improved performance
## Readability through concise and intuitive syntax
## Potentially better performance for I/O bound tasks
## Ability to represent infinite sequences
```

```python
# Q8. What is a lambda function in Python and when is it typically used?

## A lambda function is a small, anonymous function defined using the lambda keyword.
## It can have any number of arguments but can only have one expression.
## Lambda functions are often used for short, simple operation that don't require a full function definition.

add_one = lambda x: x + 1
print(add_one(5))
```

```python
# Q9. Explain the purpose and usage of the `map()` function in Python.

## The map() function applies a given function to each item of an iterable and return an iterator containing the results.
## It essentially "maps" the function to each element of the iterable .

# syntax map(function, iterable)

numbers = [1, 2, 3, 4, 5]
squared = map(lambda x: x**2, numbers)
print(list(squared))
```
